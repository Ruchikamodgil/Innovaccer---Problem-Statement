```python
X=df.iloc[:,3:35]
```

```python
Y=df.iloc[:,2]
```

```python
Y
```

```python
X["Lymph_Node_Status"].replace({"?":"0"},inplace=True)
X
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X,Y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(32).plot(kind='barh')
plt.show()
```

# Normalisation

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X= sc.fit_transform(X)
```

```python
X
```

# Train - Test - Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,random_state = 0)
```

# Model Training
Regression model to predict the recurrence time for patients whose outcome is R.
1. LinearRegression
2. RandomForest

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
model_score = model.score(X_train,Y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
print( np.sqrt(((y_predicted - Y_test) ** 2).mean())) #root mean squared error of linear model
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 35, random_state = 0,criterion='mse', max_depth=9)
regressor.fit(X_train, Y_train)
```

```python
y_predicted = regressor.predict(X_test)
```

```python
print( np.sqrt(((y_predicted - Y_test) ** 2).mean())) #root mean squared error of random forest model
```

```python
#As error generated from regressor model is comapritively smaller
pred = regressor.predict(X_test)
print("Predicted Values")
print(pred)
print("Actual Values")
x=Y_test
print(x)
```
